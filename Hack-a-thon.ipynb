```python
df_slythindor.set_index(df_slythindor['date'], drop=True, inplace=True)
```

```python
# Filter to Type A
typeA_data = df_slythindor[df_slythindor.store_type == 'A']


# Plot the sales over time
typeA_data[['weekly_sales']].plot()
```

```python
#Store 1 data autocorrelation for week-lag periods, year 2010
print typeA_data.weekly_sales.autocorr(lag=1) # -0.12
print typeA_data.weekly_sales.autocorr(lag=2) # -0.03
```

```python
# ARMA
from pandas.tools.plotting import autocorrelation_plot
autocorrelation_plot(typeA_data.weekly_sales);
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(typeA_data.weekly_sales, lags=100);
```

```python
from statsmodels.tsa.arima_model import ARMA

typeA_sales_data = typeA_data[['weekly_sales']]
model = ARMA(typeA_sales_data, (2, 2)).fit()
model.summary()
```

```python
model.resid.plot();
```
